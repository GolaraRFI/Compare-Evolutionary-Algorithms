```
import numpy as np
from numpy import ones
import matplotlib.pyplot as plt
import math
import seaborn as sns
import random

sns.set(style='darkgrid')
sns.set(font_scale=1)
plt.rcParams["figure.figsize"] = (12,6)
```

```
def rastrigin (x):
  x_pow = x**2
  t1 = 10 * (x_pow)
  c1 = (x<-5.12)+(x>5.12).astype(np.int8)
  t2 = (x_pow)-10*np.cos(2*np.pi*x)
  c2 = 1-c1
  t = (t1*c1)+(t2*c2)
  return t.sum(axis=-1)+10*t.shape[-1]

def schwefel (x):
  t1 = 0.02*(x**2)
  c1 = (x<-500)+(x>500).astype(np.int8)
  t2 = -x*np.sin(np.sqrt(np.abs(x)))
  c2 = 1-c1
  t = (t1*c1)+(t2*c2)
  return t.sum(axis=-1)+418.9829*t.shape[-1]
 
def Griewank(X):
  Fit_list = []
  for i in range(len(X)):
    SUM = 0
    DOT = 1
    for j in range(len(X[i])):
      SUM = SUM + (X[i][j]**2/4000)
      DOT = DOT * math.cos(X[i][j] / math.sqrt(j+1))
      j+=1
    FIT_0 = 1 + SUM - DOT
    Fit_list.append(FIT_0)
    i+=1
  return Fit_list
```

```
x = np.arange(-5,5,0.1)
x = x.reshape(-1,1)
y = rastrigin(x)

plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.title("Rastrigin")
plt.xlabel('X.')
plt.ylabel('Y')
plt.show()
```

```
x = np.arange(-490,490,0.1)
x = x.reshape(-1,1)
y = schwefel(x)

plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.title("Schwefel")
plt.xlabel('X.')
plt.ylabel('Y')
plt.show()
```

```
X = np.arange(-100,100,0.1)
X = X.reshape(-1,1)
y = Griewank(X)

plt.figure(figsize=(10,5))
plt.plot(X,y)
plt.title("Griewank")
plt.xlabel('X.')
plt.ylabel('Y')
plt.show()
```
